```python
#Inciso A
import numpy as np
import matplotlib.pyplot as plt
N = 100000
#gerar N amostras uniformes
y = np.random.uniform(0, 1, N)
#Gerar as amostras de g(x)
yexp = y**(1/3)
#print(yexp)
#calcular a integral pelo método da importancia calculado analiticamente
integral = 1/3 * np.sum((1-yexp**2)**(3/2)/yexp**2) / N
print(f'Integral = {integral:.4f}')
#plotar o grafico com o limite inferior e superior
x = np.linspace(0, 1, N)
#curva da função original
curve_y = (1-x**2)**(3/2)
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.text(0.5, 0.8, f'Integral = {integral:.4f}', fontsize=12, color='red')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#inciso b
N = 1000000
#gerar N amostras uniformes
y = np.random.uniform(0, 1, N)
#Gerar as amostras de g(x)
yexp = np.log(np.exp(-2) + y * (np.exp(2) - np.exp(-2)))
print(yexp)
#calcular a integral pelo método da importancia calculado analiticamente
integral = (np.exp(2) - np.exp(-2)) * np.sum(np.exp(yexp**2)) / N
print(f'Integral = {integral:.4f}')
#plotar o grafico com o limite inferior e superior
x = np.linspace(-2, 2, N)
#curva da função original
curve_y = np.exp(x + x**2)
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.text(-0.5, 100, f'Integral = {integral:.4f}', fontsize=12, color='red')
plt.xlim(-2, 2)
plt.ylim(0, 400)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import math
#inciso c
N = 70
#gerar N amostras uniformes
y = np.random.uniform(0, 1, N)
#Gerar as amostras de g(x)
ang = y * (math.pi / 2)
yexp = np.tanh(ang)
print(yexp)
#calcular a integral
integral = math.pi/2 * np.sum(yexp/(1 + yexp**2)) / N
print(f'Integral = {integral:.4f}')
#plotar o grafico com o limite inferior e superior
x = np.linspace(0, 4, N)
#curva da função original
curve_y = x/((1+x**2)**2)
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.text(2, 0.4, f'Integral = {integral:.4f}', fontsize=12, color='red')
plt.xlim(0, 4)
plt.ylim(0, 0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
