```python
import numpy as np
import matplotlib.pyplot as plt
#Inciso A
N=1000000
lim_sup= 1
lim_inf= 0
# geramos valores entre os limites
y=np.random.uniform(lim_inf,lim_sup,N)
# calculamos a integral dada pela suma
integralA = sum((1 - y**2)**(3/2))/N
print(integralA)
```

```python
# Definimos a função original
def f(x):
  return (1 - x**2)**(3/2)
# Criamos os valores do eixo x
x = np.linspace(0, 1, 100)
# traçamos função
plt.plot(x, f(x))
# Fill área sob a curva
plt.fill_between(x, f(x), color='lightblue')
# Adicionamos o rótulo com o valor da área
plt.text(0.7, 0.8,f'Area = {integralA:.3f}', fontsize=12)
# rotulo e título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of (1 - x**2)**(3/2) with Area Label')
# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#Inciso B
N=1000000
#geramos valores entre os limites
y=np.random.uniform(0,1,N)
#calculamos a integral dada pela suma
integral=4*sum(np.exp((4*y)-2+(4*y-2)**2))/N
print(integral)
```

```python
# Definimos a função original
def f(x):
  return np.exp(x+x**2)
# Criamos os valores do eixo x
x = np.linspace(-2, 2, 100)
# traçamos função
plt.plot(x, f(x))
# Fill área sob a curva
plt.fill_between(x, f(x), color='lightblue')
# Adicionamos o rótulo com o valor da área
plt.text(0, 100,f'Area = {integral:.3f}', fontsize=12)
# rotulo e título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of exp(x+x**2) with Area Label')
# Show the plot
plt.show()
```

```python
#Inciso C
N=10000
#geramos valores entre os limites
y=np.random.uniform(0,1,N)
#calculamos a integral dada pela suma
integralC=sum((((1/y)-1)*(1+((1/y)-1)**2)**-2)/y**2)/N
print(integralC)
```

```python
# Definimos a função original
def f(x):
  return (x/(1+x**2)**2)
# Criamos os valores do eixo x
x = np.linspace(0, 10, 100)
# traçamos função
plt.plot(x, f(x))
# Fill área sob a curva
plt.fill_between(x, f(x), color='lightblue')
# Adicionamos o rótulo com o valor da área
plt.text(4, 0.05,f'Area = {integralC:.3f}', fontsize=12)
# rotulo e título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of (x/(1+x**2)**2) with Area Label')
# Show the plot
plt.show()
```
